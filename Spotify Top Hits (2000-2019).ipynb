## Compreender o conjunto de dados.

O conjunto de dados é composto por 2000 filas e 18 colunas. Assim, podemos dizer que este conjunto de dados contém as 2000 canções do Top 2000 do período analisado. A fim de aumentar a profundidade de compreensão dos dados que temos, temos de compreender o conteúdo de cada coluna presente no conjunto de dados. A descrição das colunas é a seguinte :

**artist:** Nome do artista.

**song:** Nome da faixa.

**duration_ms:** Duração da faixa em milissegundos.

**explicit:** Se o conteúdo de uma canção é ou não ofensivo ou inapropriado para crianças.

**year:** Ano de lançamento da faixa.

**popularity:** Quanto mais alto for o valor, mais popular é a canção.

**danceability:** Descreve quão adequada a faixa é para dançar com base no ritmo, estabilidade do ritmo, força da batida, e outras, onde um valor mais próximo de 0,0 indica uma dançarabilidade inferior e um valor mais próximo de 1,0 indica uma dançarabilidade superior.

**energy:** representa a medida perceptual de intensidade e actividade indo de 0,0 a 1,0

**key:** Em que chave está a pista. Utiliza o inteiro para representar chaves onde 0 = C, 1 = D♭/D♭, 2 = D, e assim por diante. No caso de não ser detectada nenhuma chave, o valor é -1.

**loudness:** Representa o ruído global de uma faixa em decibéis (dB) e é a qualidade de um som que é o principal correlato psicológico da força física (amplitude). Os valores da sonoridade são calculados em média em toda a faixa, com valores tipicamente entre -60 e 0 dB.

**mode:** Indica se a faixa está numa escala maior ou menor, onde maior = 1 e menor = 0.

**speechiness:** Detecta a presença de palavras faladas numa pista. Quanto mais exclusivamente a fala - como a gravação, por exemplo um talk show ou um audiolivro, mais próximo de 1,0 será o valor atribuído. Valores acima de 0,66 descrevem faixas que são na sua maioria feitas inteiramente de palavras faladas, enquanto valores entre 0,33 e 0,66 descrevem faixas que podem conter tanto música como fala. Os valores abaixo de 0,33 representam muito provavelmente música e outras faixas não faladas.

**acousticness:** Uma medida de confiança de 0,0 a 1,0 que indica se uma faixa é acústica ou não, onde 1,0 representa uma confiança elevada a faixa é acústica.

**instrumentalness:** Prevê se a faixa contém ou não vozes. Quanto mais próximo o seu valor for de 1,0, maior é a probabilidade de a faixa não conter vocais. Qualquer valor acima de 0,5 destina-se a representar faixas instrumentais, mas a confiança é maior à medida que o valor se aproxima de 1,0. As faixas "Ohh" e "aah" são tratadas como instrumentais neste contexto.

**liveness:** Detecta a presença de um público na gravação. Um valor acima de 0,8 indica maior probabilidade de a faixa ter sido executada ao vivo.

**valence:** Uma medida de 0,0 a 1,0 que descreve a positividade transmitida pela pista. Faixas com valência mais alta podem desencadear emoções positivas, tais como felicidade, alegria e euforia, enquanto faixas com valência mais baixa podem desencadear emoções negativas, tais como tristeza, depressão e raiva.

**tempo:** Tempo em batidas por minuto (BPM). Em terminologia musical, tempo é a velocidade ou ritmo de uma determinada peça e deriva directamente da duração média da batida.

**genre:** Género da pista.

## Verificar os anos de lançamento

```python
dados_Spotify.year.unique()
```

Embora o título do _dataset_ diga que contém êxitos de 2000 a 2019, podemos ver que algumas faixas têm o seu ano de lançamento datado de 1998, 1999 e 2020. Vamos ver quantas faixas foram lançadas nesses três anos.

```python
# Print de quantas canções lançadas em 1998, 1999 e 2020.
print(len(dados_Spotify.query("year == 1998")))
print(len(dados_Spotify.query("year == 1999")))
print(len(dados_Spotify.query("year == 2020")))
```

Temos 1 música de 1998, 38 músicas de 1999 e 3 músicas de 2020.
Considerando que o objetivo deste estudo é analisar as músicas desde 2000 até 2019, vamos retirar do conjunto de dados todas as músicas lançadas em qualquer ano que esteja fora do período destinado a ser analisado.

```python
# Armazenar todas as músicas lançadas antes de 2000 e depois de 2019 num novo dataframe.
dados_Spotify_years_drop = dados_Spotify[(dados_Spotify['year'] <2000) | (dados_Spotify['year'] > 2019)].index

# Remover as músicas de 1998, 1999 e 2020 do dataframe original.
dados_Spotify = dados_Spotify.drop(dados_Spotify_years_drop)

# Verificar os valores dos anos de lançamento do dataframe.
dados_Spotify.year.unique()
```
